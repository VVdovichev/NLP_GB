```python
import pandas as pd
import re
import pickle
import nltk
from functools import partial
from string import punctuation
```

```python
train_df = pd.read_csv('train_tweets.csv')
test_df = pd.read_csv('test_tweets.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
data = pd.concat([train_df, test_df]).reset_index(drop=True)
data.head()
```

```python
data['clean_tweets'] = data['tweet'].copy()
```

```python
data.head()
```

1. Удалим @user из всех твитов с помощью паттерна "@[\w]*". Для этого создадим функцию: ● для того, чтобы найти все вхождения паттерна в тексте, необходимо использовать re.findall(pattern, input_txt) ● для для замены @user на пробел, необходимо использовать re.sub()

```python
data['clean_tweets'] = data['clean_tweets'].str.replace(r'@\w*', '', regex=True)
```

```python
data.head()
```

2. Изменим регистр твитов на нижний с помощью .lower().

```python
data['clean_tweets'] = data['clean_tweets'].str.lower()
data.head()
```

3. Заменим сокращения с апострофами (пример: ain't, can't) на пробел, используя apostrophe_dict. Для этого необходимо сделать функцию: для каждого слова в тексте проверить (for word in text.split()), если слово есть в словаре apostrophe_dict в качестве ключа (сокращенного слова), то заменить ключ на значение (полную версию слова).

```python
def remap_func(text, dictionary):
    return ' '.join([dictionary.get(word, word) for word in text.split()])
```

```python
data['clean_tweets'] = data['clean_tweets'].apply(partial(remap_func, dictionary=apostrophe_dict))
data.head()
```

4.  Заменим сокращения на их полные формы, используя short_word_dict. Для этого
воспользуемся функцией, используемой в предыдущем пункте.


```python
data['clean_tweets'] = data['clean_tweets'].apply(partial(remap_func, dictionary=short_word_dict))
data.head()
```

5. . Заменим эмотиконы (пример: ":)" = "happy") на пробелы, используя emoticon_dict.
Для этого воспользуемся функцией, используемой в предыдущем пункте.


```python
data['clean_tweets'] = data['clean_tweets'].apply(partial(remap_func, dictionary=emoticon_dict))
data.head()
```

6. Заменим пунктуацию на пробелы, используя re.sub() и паттерн r'[^\w\s]'.
7. Заменим спец. символы на пробелы, используя re.sub() и паттерн r'[^a-zA-Z0-9]'.
8. Заменим числа на пробелы, используя re.sub() и паттерн r'[^a-zA-Z]'.


```python
data['clean_tweets'] = data['clean_tweets'].str.replace(f'[{punctuation}0-9]', ' ', regex=True)
data.head()
```

9. Удалим из текста слова длиной в 1 символ, используя ' '.join([w for w in x.split() if
len(w)>1]).

```python
data['clean_tweets'] = data['clean_tweets'].apply(lambda text: ' '.join([w for w in text.split() if len(w) > 1]))
data.head()
```

10. Поделим твиты на токены с помощью nltk.tokenize.word_tokenize, создав новый
столбец 'tweet_token'.


```python
data['tweet_token'] = data['clean_tweets'].apply(nltk.tokenize.word_tokenize)
data.head()
```

11. Удалим стоп-слова из токенов, используя nltk.corpus.stopwords. Создадим столбец
'tweet_token_filtered' без стоп-слов.

```python
stop_words = nltk.corpus.stopwords.words('english')
data['tweet_token_filtered'] = data['tweet_token'].apply(lambda tokens: [word for word in tokens if word not in stop_words])
data.head()
```

12. Применим стемминг к токенам с помощью nltk.stem.PorterStemmer. Создадим
столбец 'tweet_stemmed' после применения стемминга.


```python
stemmer = nltk.stem.PorterStemmer()
data['tweet_stemmed'] = data['tweet_token'].apply(lambda tokens: [stemmer.stem(word) for word in tokens])
data.head()
```

13.  Применим лемматизацию к токенам с помощью
nltk.stem.wordnet.WordNetLemmatizer. Создадим столбец 'tweet_lemmatized' после
применения лемматизации.


```python
lemma = nltk.stem.wordnet.WordNetLemmatizer()
data['tweet_lemmatized'] = data['tweet_token'].apply(lambda tokens: [lemma.lemmatize(word) for word in tokens])
data.head()
```

14. Сохраним результат предобработки в pickle-файл.

```python
pickle.dump(data['tweet_lemmatized'].to_dict(), open('tweet_lemmatized.pkl', 'wb'))
```
